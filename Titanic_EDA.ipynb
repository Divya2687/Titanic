```python
def avg_age_wrt_title(x):
    '''avg of the age wrt the titles'''
    return(round(df[(df['Title']==x)]['Age'].mean(),2))
```

```python
avg_age_wrt_title('Master')
```

```python
avg_age_wrt_title('Dr')
```

```python
for  i in ['Capt','Col','Don','Dr','Jonkheer','Lady','Major','Master','Miss','Mlle','Mme','Mr','Mrs','Ms','Rev','Sir','theCountess']:
    print(i, ':', avg_age_wrt_title(i))
```

```python
df.loc[(df['Age'].isnull() & df['Title'] == 'Capt'),'Age']=70
df.loc[(df['Age'].isnull() & df['Title'] == 'Col'),'Age']=58
df.loc[(df['Age'].isnull() & df['Title'] == 'Don'),'Age']=48
```

```python
title_mean = []
for  i in ['Capt','Col','Don','Dr','Jonkheer','Lady','Major','Master','Miss','Mlle','Mme','Mr','Mrs','Ms','Rev','Sir','theCountess']:
    title_mean.append((i,  avg_age_wrt_title(i)))
```
